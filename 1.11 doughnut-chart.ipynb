```python
from matplotlib import pyplot as plt
```

```python
fruit = ["apple", "banana", "mango", "orange"]
quantity = [55,10,20,100]
```

```python
plt.pie(quantity,labels =fruit,radius=2) #ye bahar wale ka radius hai
plt.pie([1],colors=["w"],radius=1)  #andar ka circle
```

adding percentage

```python
plt.pie(quantity,labels =fruit, autopct="%0.1f%%", colors=["green","red","blue","black"])
plt.show
```
