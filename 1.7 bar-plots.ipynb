when you want to understand the categorical data, we use the bar plot

```python
import numpy as np
from matplotlib import pyplot as plt




students={"muzi":5,"fara":10,"prince":15}
```

```python
names =list(students.keys())
marks =list(students.values())
```

```python
plt.bar(names, marks)
plt.show
```

```python
plt.bar(names, marks)
plt.title("Bar Plot")
plt.ylabel("marks")
plt.xlabel("Names")
plt.grid(True)
plt.show()
```

Horizontal Bar chart

```python
plt.barh(names, marks,color="g")
plt.title("Bar Plot")
plt.ylabel("marks")
plt.xlabel("Names")
plt.grid(True)
plt.show()
```
