```python
import numpy as np
from matplotlib import pyplot as plt


#define one numpy array
x= np.arange(1,11)

#define 2nd numpy array
y1= 2*x

#define 3rd numpy array
y2= 3*x
```

```python
#subplot syntax:
#########################################
#  subplot(row no.,column no., index)                                     #
#########################################


#subplot-1
plt.subplot(1,2,1)    #in one row 2 subplots and last perimeter indicates index       
plt.plot(x,y1,color= "g", linestyle= ":", linewidth=2)
plt.title("Line subplot-1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True) 


#subplot-2
plt.subplot(1,2,2) 
plt.plot(x,y2,color= "r", linestyle= "-", linewidth=2)
plt.title("Line subplot-2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)        #with graph papers
plt.show()
```
