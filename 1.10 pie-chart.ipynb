```python
from matplotlib import pyplot as plt
```

```python
fruit = ["apple", "banana", "mango", "orange"]
quantity = [55,10,20,100]
```

```python
plt.pie(quantity,labels =fruit)
plt.show
```

adding percentage

```python
plt.pie(quantity,labels =fruit, autopct="%0.1f%%", colors=["green","red","blue","black"])
plt.show
```
